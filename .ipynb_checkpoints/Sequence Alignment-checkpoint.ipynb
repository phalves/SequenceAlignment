####Task 2 with O(m.n) space and time

```python
import random
import resource
import numpy as np
import time
from math import log

z = "ABCD"
gamma = 0.7
alpha = 1


def task2():
    result = [[],[],[]]
    for i in range(8):
        print i
        t = []
        m = []
        size = 10*pow(2, i)
        for j in range(10):
            start_time = time.clock()/60
            start_mem = memory_usage()
            x = [random.choice(z) for _ in range(size)]
            y = [random.choice(z) for _ in range(size)]
            M, cost = sequence_alignemt(x, y, g, a)
            out_x, out_y = get_sequence(x, y, g, a, M)
            t.append((time.clock() - start_time)/60)
            m.append(memory_usage() - start_mem)

        result[0].append(size)
        # Saves the mean time
        result[1].append(np.median(t))
        # Saves the mean of consumed memory
        result[2].append(np.median(m))
    return result


def memory_usage():
    return resource.getrusage(resource.RUSAGE_SELF).ru_maxrss / 10**3
```

```python
result = task2()
```

```python
import matplotlib.pyplot as plt


plt.figure(1)
# Size X Time
plt.plot(result[1], result[0])
plt.xlabel(u"Tempo médio (segundos)")
plt.ylabel(u"Tamanho das cadeias")

# Size i X Memory
plt.figure(2)
plt.plot(result[2], result[0])
plt.xlabel(u"Consumo médio de Memória (kB)")
plt.ylabel(u"Tamanho das cadeias")
plt.show()
```

```python
def divide_and_conquer_alignment(x, y, g, a):
    m = len(x)
    n = len(y)
    if m <= 2 or n <= 2:
        print("mn ", m, n)
        v, _ = linear_sequence_alignment(x, y, g, a)

    if m > 0 and n > 0:
        _f, _ = linear_sequence_alignment(x, y[:int(n/2 - 1)], g, a)
        _g, _ = linear_sequence_alignment(x, y[int(n/2):], g, a)
        r = [_f[i] + _g[i] for i in range(len(_f))]
        q = r.index(min(r))

        divide_and_conquer_alignment(x[:(q - 1)], y[:int(n/2 - 1)], g, a)
        divide_and_conquer_alignment(x[q:(n-1)], y[int(n / 2):], g, a)
```

```python
Divide-and-Conquer-Alignment( X , Y );
    Let m be the number of symbols in X
    Let n be the number of symbols in Y
    If m ≤ 2 or n ≤ 2 then
        Compute optimal alignment using Alignment( X , Y )
        Let q be the ...

    Call Space-Efficient-Alignment( X , Y[1 : n/2] )
    Call Backward-Space-Efficient-Alignment( X , Y[n/2 + 1 : n] )
    Let q be the index minimizing f (q, n/2) + g(q, n/2)
    Add (q, n/2) to global list P
    Divide-and-Conquer-Alignment( X[1 : q] , Y[1 : n/2] )
    Divide-and-Conquer-Alignment( X[q + 1 : n] , Y[n/2 + 1 : n] )
    Return P
```
